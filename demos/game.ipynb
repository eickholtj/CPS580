## Learning the to Identify the "Winner" of a Game ##

Two players pick a value between 0 and 1.  The player that picks the higher value wins!

```
import urllib

import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers

# Read in the data from a URL; works for small datasets
url = "https://github.com/eickholtj/CPS580/raw/master/data/game-data.txt"
file = urllib.request.urlopen(url)
data = np.loadtxt(file)

print("Let's have a look at the data...")
print(data[:10,:])

# Break of the targets from the input
X = data[:,0:2]
Y = data[:,2]

# We will pull of the last 20% as a test set
N = data.shape[0]
split = int(N * 0.8)
train_X = X[0:split,:]
train_Y = Y[0:split]
test_X = X[split:,:]
test_Y = Y[split:]

print("Shape of training input: ", train_X.shape)
print("Shape of training targets: ", train_Y.shape)
print("Shape of test input: ", test_X.shape)
print("Shape of test targets: ", test_Y.shape)

# create model
model = Sequential()
model.add(Dense(2, input_dim=2, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))

sgd = optimizers.SGD(lr=0.0001, momentum=0.9)
model.compile(loss='mean_squared_error', optimizer=sgd, metrics=['accuracy'])

model.fit(train_X, train_Y, epochs=15, batch_size=50)
scores = model.evaluate(test_X, test_Y)

#raw_preds = model.predict(test_X)

# Print out the loss and the accuracy
print(scores)



```
